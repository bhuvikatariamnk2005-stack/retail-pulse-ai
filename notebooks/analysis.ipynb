# Retail Pulse - Data Analysis Project

This project analyzes retail sales data to generate business insights, customer segmentation, and sales forecasting using Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/retail_pulse_cleaned.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df["customer_id"] = np.random.randint(1000, 2000, size=len(df))
df["order_date"] = pd.date_range(start="2023-01-01", periods=len(df), freq="D")
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df.groupby("category")["sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby("region")["profit"].sum().plot(kind="bar")
plt.title("Profit by Region")
plt.show()
```

## Key Insights

- Certain categories dominate total sales
- Region-wise performance varies significantly
- Discounts impact profitability
- Few products contribute most revenue (80/20 rule)
